```python
import csv
import datetime
import pandas as pd

df = pd.read_csv(
    'in_lang_docs_all_bibitems.tsv',
    header=None,
    names = ['lang', 'uuid', 'citing_mid', 'cited_mid', 'citing_aid', 'cited_aid', 'bibitem_str'],
    sep='\t',
    low_memory=False,
    quoting=csv.QUOTE_NONE
)
```

```python
unmatched_lang_counts = df[df.lang.notnull() & df.cited_mid.isnull()].lang.value_counts().to_dict()
for lang, total_count in df[df.lang.notnull()].lang.value_counts().to_dict().items():
    print('{}: {}/{} ({:.4f})'.format(
        lang,
        unmatched_lang_counts[lang],
        total_count,
        unmatched_lang_counts[lang]/total_count
    ))
# df[df.lang == 'english'].bibitem_str.to_list()  # who'd've guessed there's also "(in English)" in English papers
```

```python
df_ja_w_link = df[(df.lang == 'japanese') & df.bibitem_str.str.contains('http')]
print('--- matched ---')
for i, rovv in enumerate(df_ja_w_link[df_ja_w_link.cited_mid.notnull()].iterrows()):
    row = rovv[1]
    print('[{}] {}->{} ({})\n{}\n'.format(
        i,
        row.citing_aid,
        int(row.cited_mid),
        row.uuid,
        row.bibitem_str.replace('¦', '')))
print('--- not matched ---')
for i, rovv in enumerate(df_ja_w_link[df_ja_w_link.cited_mid.isnull()].iterrows()):
    row = rovv[1]
    print('[{}] {}->? ({})\n{}\n'.format(
        i,
        row.citing_aid,
        row.uuid,
        row.bibitem_str.replace('¦', '')))
```
